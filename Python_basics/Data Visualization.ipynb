### Make a plot using ggplot where x = pop and y = lifeExp

```python
(ggplot(my_file) + aes(x='pop' , y='lifeExp') + geom_point())
```

```python
(ggplot(my_file) + aes(x='pop', y='lifeExp', color = 'continent') + geom_point())
```

```python
(ggplot(my_file) + aes(x='pop', y='lifeExp', color = 'year') + geom_point())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', color = 'continent', by = 'country') + geom_line())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp') + geom_line(aes(color = 'continent')) + geom_point())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', by = 'country') + geom_line(aes(color = 'continent')) + geom_point())
```
